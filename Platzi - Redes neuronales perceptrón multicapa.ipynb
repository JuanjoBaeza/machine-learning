```python
import numpy as np
from matplotlib import pyplot as plt
import time
```

```python
plt.figure(figsize=(12,4))
x=np.linspace(-10,10,50)

plt.subplot(1,2,1)
plt.plot(x,1/(1+np.e**(-x)))
plt.title('Funcion sigmoide')
plt.grid()

plt.subplot(1,2,2)
plt.plot(x,np.tanh(x))
plt.title('Funcion Tangente Hiperb√≥lica')
plt.grid()
plt.show()
```
