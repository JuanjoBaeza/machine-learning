```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# El campo edad = edad del cliente, cantidad = cantidad de personas que iban con esa persona, la columna vehiculo = si llegÃ³ o no en un vehÃ­culo particular, la columna pago se refiere a su forma de pago (efectivo, tarjeta de credito, dÃ©bito o vales), y por Ãºltimo la columna monto que indica cuÃ¡nto comprÃ³.
ventas = pd.read_csv("datasets/reg-lineal-ventas.csv")
objetivo = "importe"
independientes = ventas.drop(columns=['importe']).columns

modelo = LinearRegression()

X = ventas[independientes]
y = ventas[objetivo]

modelo.fit(X,y)

ventas["importe_prediccion"] = modelo.predict(X)
preds = ventas[["importe", "importe_prediccion"]].head(50)

# persona de 41 aÃ±os, solo, en su propio vehÃ­culo y pagando en efectivo
talvez = modelo.predict([[47,3,1,1]])
print ("Importe de compra: ")
print (np.round(talvez,1) ,"â‚¬")

preds.plot(kind='bar',figsize=(18,8))
plt.grid(linewidth='2')
plt.grid(linewidth='2')
plt.grid(None)
plt.show()
```
