```python
#  For each token in the sentence above, print its text, POS tag, dep tag and lemma
for word in list_of_sentence[3]:
    print(word.text, word.pos_, word.dep_, word.lemma_)
```

```python
# Import the Matcher library:
from spacy.matcher import Matcher

matcher = Matcher(nlp.vocab)
```

```python
# Write a matcher called 'Swimming' that finds both occurrences of the phrase "swimming vigorously" in the text
# Create a pattern and add it to matcher:

#pattern1 = [{'LOWER':'swimming'}]
pattern2 = [{'LOWER':'swimming'}, {'IS_SPACE':True}, {'LOWER':'vigorously'}]
```

```python
matcher.add("Swimming", [pattern2])
```

```python
found_match = matcher(doc) 
```

```python
found_match
```

```python
for match_id, start, end in found_match:
    string_id = nlp.vocab.strings[match_id]
    span = doc[start-9:end+14]
    
    print(span.text)
```
