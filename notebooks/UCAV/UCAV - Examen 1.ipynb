```python
def rangos_edad(d):

    datos=d['data']

    rangos=[]

    for m in datos:

        dato=m["rango_edad_de_la_madre"]

        if not (dato in rangos):

            rangos.append(dato)

    return rangos

rangos_edad(d)
```

```python
def nacimientos_mujeres_15(d):

    datos=d['data']

    nac=0

    for m in datos:

        if m["rango_edad_de_la_madre"]=="Menores de 15 años":

            nac=nac+m['numero_nacimientos']

    return nac

nacimientos_mujeres_15(d)
```

```python
def nacimientos_por_localidad(d):

    datos=d['data']

    nacimientos={}

    for m in datos:

        if m['numero_nacimientos']>0:       
    
            dato=m["municipio_nombre"]
            claves=nacimientos.keys()
            
            if not (dato in claves):

                nacimientos[dato]=m['numero_nacimientos']

            else:

                nacimientos[dato]=nacimientos[dato]+m['numero_nacimientos']

    return nacimientos

nacimientos_por_localidad(d)
```

```python
def nacimientos_por_localidad_sexo(d):

    datos=d['data']

    nacimientos={}

    for m in datos:

        if m['numero_nacimientos']>0:       

            dato=m["municipio_nombre"]

            claves=nacimientos.keys()
            
            sex=m["sexo"]

            if not ((dato,sex) in claves):

                nacimientos[(dato,sex)]=m['numero_nacimientos']

            else:

                nacimientos[(dato,sex)]=nacimientos[(dato,sex)]+m['numero_nacimientos']
                

    return nacimientos

nacimientos_por_localidad_sexo(d)
```

```python
def nacimientos_por_localidad_sexo(d):

    datos=d['data']

    nacimientos={}

    for m in datos:

        if m['numero_nacimientos']>0:       

            dato=m["municipio_nombre"]

            claves=nacimientos.keys()
            
            sex=m["sexo"]

            if not ((dato,sex) in claves):

                nacimientos[(dato,sex)]=m['numero_nacimientos']

            else:

                nacimientos[(dato,sex)]=nacimientos[(dato,sex)]+m['numero_nacimientos']
                

    return sorted(nacimientos.items(), key=lambda item: item[1])

nacimientos_por_localidad_sexo(d)
```
