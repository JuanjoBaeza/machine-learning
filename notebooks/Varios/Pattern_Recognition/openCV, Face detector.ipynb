```python
#https://github.com/opencv/opencv/tree/master/data/haarcascades
#https://malinian.medium.com/face-detection-using-haar-cascade-xml-files-in-python-e85b10b41755
import cv2
from PIL import Image
from IPython.display import display

face_cascade = cv2.CascadeClassifier('../../data/datasets/face-detector.xml')

img = cv2.imread('../../data/datasets/images/face1.jpg')
faces = face_cascade.detectMultiScale(img, 1.1, 4)

for (x, y, w, h) in faces: 
  cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)
  
cv2.imwrite("../../data/datasets/images/output-file.jpg", img)

img = Image.open('../../data/datasets/images/output-file.jpg')

display(img)
```
